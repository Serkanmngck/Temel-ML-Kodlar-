```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier,export_graphviz
from sklearn.model_selection import train_test_split
import graphviz
```

```python
df=pd.read_csv("heart.csv")
df.head(3)
```

```python
df.info()
```

```python
y=df['output']
x=df.drop("output",axis=1)
```

```python
tree=DecisionTreeClassifier()
model=tree.fit(x,y)
```

```python
model.score(x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=16,train_size=0.70)
```

```python
tree=DecisionTreeClassifier()
model1=tree.fit(x_train,y_train)
model1.score(x_test,y_test)
```

```python
dot=export_graphviz(model1)
gorsel=graphviz.Source(dot)
gorsel
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest=RandomForestClassifier()
model=forest.fit(x,y)
model.score(x,y)
```

```python
forest=RandomForestClassifier(n_estimators=250,max_depth=5)  ##n_estimators aa says sistemi etikler, max depth derinlik ezberi etkiler 
model=forest.fit(x_train,y_train)
model.score(x_test,y_test)
```
