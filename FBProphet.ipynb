```python
import pandas as pd
import numpy as np
from prophet import Prophet
import yfinance as yf
```

```python
df=yf.download('BTC-USD',"2015-01-01")
```

```python
df=df['Close']
df=df.reset_index()
df.head(3)
```

```python
df.columns=['ds','y']
df.head(3)
```

```python
model=Prophet()
model.fit(df)
```

```python
gelecek=model.make_future_dataframe(360)
tahmin=model.predict(gelecek)
```

```python
model.plot(tahmin)
```

```python
model.plot_components(tahmin);
```
